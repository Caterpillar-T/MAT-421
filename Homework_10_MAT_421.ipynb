**Section 22.1: ODE Initial Value Problem Statement**

Ordinary differential equations, sometimes called ODE's, are differential equations where the independent variable and derivatives are in the same one dimension.

An ODE can be written $F(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2}, ... , \frac{d^{n-1}f(x)}{dx^{x-1}}) = \frac{d^nf(x)}{dx^n}$. F is an arbitary function and $n$ is the oder of the differential equation.

Unlike ODE's, PDE's are general forms of differential equations where partial derivatives are not dependent on order.

**Section 22.2: Reduction of Order**

It is common practice to reduce the order of a differential equation to first order in effort to make a high order differential equation more solvable.

Which can be expressed $f^{(n)}(t)=F(t,f(t),f^{(1)}(t), ... , f^{(n-1)}(t))$.

And to take the state to be
$$
S(t)=
\begin{bmatrix}
f(t) \\
f^{(1)}(t) \\
f^{(2)}(t) \\
f^{(3)}(t) \\
\vdots \\
f^{(n-1)}(t)
\end{bmatrix}
$$

**Section 22.3: The Euler Method**

The Explicit Euler Formula, written $S(t_{j+1}) = S(t_j) + hF(t_j,S(t_j))$, allows for computing an approximation for the state of $S(t_{j+1})$.

Now, using Python we will approximate the solution of the differential equation $\frac{df(t)}{dt} = -2f(t)$ with initial condition $f(0)=1$ which has the exact solution $f(t) = e^{-2t}$ and compare.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Define parameters
f = lambda t, s: -2 * s
h = 0.1  # Step size
t = np.arange(0, 1 + h, h)  # Numerical grid
s0 = 1  # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Exact solution
exact_solution = np.exp(-2 * t)

# Plot results
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate (Euler)')
plt.plot(t, exact_solution, 'g', label='Exact Solution')
plt.title('Approximate and Exact Solution for df/dt = -2f')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
