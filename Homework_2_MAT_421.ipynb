**Section 17.1: Interpolation Problem Statement**

Interpolation is a method by which we find an estimation function $\hat y(x)$ such that $\hat y(x_i)=y_i$

We can interpolate a function, given an $x_1$, using $\hat y(x_1)$. In this case we would call $\hat y(x)$ the interpolation function.

The convenient thing about interpolation is that no underlying model is required for the data.

**Section 17.2: Linear Interpolation**

Linear interpolation can be defined/represented at $x$ as $\hat y(x) = y_i + \frac{(y_{i+1}-y_i)(x-x_i)}{(x-{i+1}-x_i)}$

In linear interpolation, whatever point is being estimated is assumed to lie on the line joining the nearest points to the left and right

For example, if we have points $x = [7, 8, 9]$ and $y = [10, 7, 9]$ and we are looking for the linear interpolation at $x = 7.5$

Then from our function above we'd get: $\hat y(x) = 10 + \frac{(7-10)(7.5-7)}{(8-7)} = 8.5$

Which we can of course double check with Python

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [7, 8, 9]
y = [10, 7, 9]

f = interp1d(x, y)
y_hat = f(7.5)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(7.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 7.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Section 17.3: Cubic Spline Interpolation**

Cubic spline interpolation made up of a set of piecewise cubic function.

It is assumed that points $(x_i, y_i)$ and $(x_{i+1},y_{i+1})$ are joined by a cubic polynomial $S_i(x) = a_ix^3 + b_ix^2 + c_ix + d_i$.

In order to find the cubic spline interpolation function the coefficients for each cubic function must first be determined. Each cubic function required four coefficients and there are $n - 1$ cubic function for $n$ number of points.

In other words, that would mean we have to find $n - 1$ four times, or we need to find $4(n-1)$ unknowns.

For example, if we use the same points for x and y as above. $x = [7, 8, 9]$ and $y = [10, 7, 9]$ for $7 \leq x \leq 9$, we can have Python do the heavy lifting for us rather than determine the cublic spline interpolation by hand.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [7, 8, 9]
y = [10, 7, 9]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(5, 10, 50)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
